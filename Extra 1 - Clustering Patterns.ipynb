```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import yfinance
```

```python
%config Completer.use_jedi = False
```

```python
ticker = yfinance.Ticker('BTC-USD')
```

```python
df = ticker.history(period='max')
```

```python
df = df[['Close']].copy()
```

```python
df
```

```python
for i in range(1, 11):
    df['shift'+str(i)] = df['Close'].shift(i)
```

```python
df.head(20)
```

```python
df.tail()
```

```python
df = df[df.columns[::-1]].copy()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df = df.div(df['shift10'], axis=0).drop(['shift10'], axis=1)
```

```python
df
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = KMeans(n_clusters=10)
```

```python
cluster.fit(df)
```

```python
cluster.predict(df)
```

```python
result = df.copy()
```

```python
result['cluster'] = cluster.predict(df)
```

```python
result
```

```python
cluster.cluster_centers_
```

```python
fig, ax = plt.subplots(3, 4, figsize=(20,10))

for i in range(10):
    ax[i // 4][i % 4].plot(cluster.cluster_centers_[i])
    ax[i // 4][i % 4].title.set_text('Cluster ' + str(i))
```

```python
sse = list()
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    sse.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), sse)
plt.grid(True)
plt.title('Elbow curve')
```

```python
cluster = KMeans(n_clusters=15)
cluster.fit(df)
```

```python
cluster.cluster_centers_
```

```python
fig, ax = plt.subplots(5, 3, figsize=(20,20))

for i in range(15):
    ax[i // 3][i % 3].plot(cluster.cluster_centers_[i])
    ax[i // 3][i % 3].title.set_text('Cluster ' + str(i))
```

```python
result = df.copy()
result['cluster'] = cluster.predict(df)
```

```python
result[result['cluster'] == 2].sample(11).drop(['cluster'], axis=1).T.plot(legend=False)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(df, cluster.predict(df))
```

```python
score = list()
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df)
    score.append(silhouette_score(df, kmeans.predict(df)))

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), score)
plt.grid(True)
```
